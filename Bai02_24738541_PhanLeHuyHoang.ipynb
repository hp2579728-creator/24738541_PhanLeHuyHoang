````python
books = [
     {"id": 1, "title": "Lập trình Python cơ bản", "author": "Nguyễn Văn A", "year": 2020, "status": "available"},
    {"id": 2, "title": "Khoa học dữ liệu với Python", "author": "Trần Thị B", "year": 2021, "status": "borrowed"},
    {"id": 3, "title": "Machine Learning cơ bản", "author": "Lê Văn C", "year": 2022, "status": "available"},
    {"id": 4, "title": "Python nâng cao", "author": "Nguyễn Văn A", "year": 2023, "status": "available"},
    {"id": 5, "title": "Big Data với Python", "author": "Phạm Thị D", "year": 2024, "status": "borrowed"}
]

def display_all_books():
    """
    Hiển thị danh sách tất cả các sách trong thư viện
    In ra: ID, Tên sách, Tác giả, Năm xuất bản, Trạng thái
    """
    # code của bạn
    print(f"\n{'ID':<5} | {'Tên sách':<30} | {'Tác giả':<20} | {'Năm':<6} | {'Trạng thái':<10}")
    print("-" * 85)
    for b in books:
        print(f"{b['id']:<5} | {b['title']:<30} | {b['author']:<20} | {b['year']:<6} | {b['status']:<10}")
````

````python
def add_new_book(title, author, year):
    """
    Thêm sách mới vào thư viện
    - Tự động tạo ID
    - lấy giá trị truyền vô: title, author, year
    - Mặc định status = "available"
    """
    print("\n--- THÊM SÁCH MỚI ---")
    # code của bạn
    global books 
    
    # Tự động tạo ID: Lấy ID lớn nhất hiện tại cộng thêm 1
    if not books:
        new_id = 1
    else:
        new_id = max([b['id'] for b in books]) + 1
    
    new_book = {
        "id": new_id,
        "title": title,
        "author": author,
        "year": year,
        "status": "available"
    }
    
    books.append(new_book)
    print(f"✓ Đã thêm thành công sách: {title} (ID: {new_id})")

````

````python
def borrow_book(ID):
    """
    Mượn sách từ thư viện
    - Nhập ID sách cần mượn
    - Kiểm tra status = "available"
    - Cập nhật status = "borrowed"
    - Thông báo kết quả
    """
    print("\n--- MƯỢN SÁCH ---")
    
    # code của bạn
    for b in books:
        if b['id'] == ID:
            if b['status'] == "available":
                b['status'] = "borrowed"
                print(f"✓ Bạn đã mượn thành công cuốn: {b['title']}")
            else:
                print("✗ Sách này hiện đã có người mượn.")
            return
    print("✗ Không tìm thấy ID sách này.")
    
````

````python
def return_book(ID):
    """
    Trả sách cho thư viện
    - Nhập ID sách cần trả
    - Kiểm tra status = "borrowed"
    - Cập nhật status = "available"
    - Thông báo kết quả
    """
    print("\n--- TRẢ SÁCH ---")
    
    # code của bạn
    for b in books:
        if b['id'] == ID:
            if b['status'] == "borrowed":
                b['status'] = "available"
                print(f"✓ Cảm ơn bạn đã trả cuốn: {b['title']}")
            else:
                print("✗ Sách này hiện đang ở trong kho, không cần trả.")
            return
    print("✗ Không tìm thấy ID sách này.")
````

````python
def search_by_author():
    """
    Tìm kiếm sách theo tác giả
    - Nhập tên tác giả
    - Duyệt list và hiển thị tất cả sách của tác giả đó
    - Sử dụng lower() để không phân biệt hoa thường
    """
    print("\n--- TÌM SÁCH THEO TÁC GIẢ ---")
    
    # code của bạn 
    author_name = input("Nhập tên tác giả cần tìm: ").strip().lower()
    print(f"\n--- KẾT QUẢ TÌM KIẾM CHO TÁC GIẢ: {author_name.upper()} ---")
    found = False
    for b in books:
        if author_name in b['author'].lower():
            print(f"ID: {b['id']} | {b['title']} ({b['year']}) | [{b['status']}]")
            found = True
    if not found:
        print("Không tìm thấy cuốn sách nào của tác giả này.")

````

````python
def filter_by_year():
    """
    Lọc sách theo khoảng năm xuất bản
    - Nhập năm bắt đầu và năm kết thúc
    - Hiển thị các sách trong khoảng năm đó
    """
    print("\n--- LỌC SÁCH THEO NĂM XUẤT BẢN ---")
    
    # code của bạn
    try:
        start = int(input("Từ năm: "))
        end = int(input("Đến năm: "))
        print(f"\n--- SÁCH XUẤT BẢN TỪ {start} ĐẾN {end} ---")
        for b in books:
            if start <= b['year'] <= end:
                print(f"{b['id']}. {b['title']} - {b['author']} ({b['year']})")
    except ValueError:
        print("Vui lòng nhập năm là một số nguyên.")
````

````python
def show_statistics():
    """
    Hiển thị thống kê thư viện
    - Tổng số sách
    - Số sách đang available
    - Số sách đang borrowed
    - Tính phần trăm
    """
    print("\n--- THỐNG KÊ THƯ VIỆN ---")
    
    # code của bạn 
    otal = len(books)
    if total == 0:
        print("Thư viện hiện đang trống.")
        return
    
    avail = sum(1 for b in books if b['status'] == "available")
    borrowed = total - avail
    
    per_avail = (avail / total) * 100
    per_borrowed = (borrowed / total) * 100
    
    print(f"\n--- THỐNG KÊ THƯ VIỆN ---")
    print(f"- Tổng số sách: {total}")
    print(f"- Sẵn có: {avail} ({per_avail:.1f}%)")
    print(f"- Đang cho mượn: {borrowed} ({per_borrowed:.1f}%)")
````

````python
def main_menu():
    # code của bạn
    while True:
        print("\n" + "="*40)
        print("   HỆ THỐNG QUẢN LÝ THƯ VIỆN PYTHON")
        print("="*40)
        print("1. Hiển thị toàn bộ sách")
        print("2. Thêm sách mới")
        print("3. Mượn sách (theo ID)")
        print("4. Trả sách (theo ID)")
        print("5. Tìm sách theo tác giả")
        print("6. Lọc sách theo năm")
        print("7. Xem thống kê")
        print("0. Thoát chương trình")
        
        choice = input("\nChọn chức năng (0-7): ")
        
        if choice == "1":
            display_all_books()
        elif choice == "2":
            t = input("Nhập tên sách: ")
            a = input("Nhập tác giả: ")
            y = int(input("Nhập năm xuất bản: "))
            add_new_book(t, a, y)
        elif choice == "3":
            id_borrow = int(input("Nhập ID sách muốn mượn: "))
            borrow_book(id_borrow)
        elif choice == "4":
            id_return = int(input("Nhập ID sách muốn trả: "))
            return_book(id_return)
        elif choice == "5":
            search_by_author()
        elif choice == "6":
            filter_by_year()
        elif choice == "7":
            show_statistics()
        elif choice == "0":
            print("Tạm biệt!")
            break
        else:
            print("Lựa chọn không hợp lệ!")

if __name__ == "__main__":
        main_menu()
````
