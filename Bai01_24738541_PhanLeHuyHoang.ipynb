## **III. Data init**

````python
# Danh sách sinh viên ban đầu
students = [
    {
        "id": "SV001",
        "name": "Nguyễn Văn An",
        "age": 20,
        "scores": {"SAD": 8.5, "DSA": 7.0, "BigData": 9.0}
    },
    {
        "id": "SV002",
        "name": "Trần Thị Bình",
        "age": 21,
        "scores": {"SAD": 6.5, "DSA": 7.5, "BigData": 6.0}
    },
    {
        "id": "SV003",
        "name": "Lê Văn Cường",
        "age": 19,
        "scores": {"SAD": 9.0, "DSA": 8.5, "BigData": 9.5}
    },
    {
        "id": "SV004",
        "name": "Phạm Thị Dung",
        "age": 20,
        "scores": {"SAD": 5.0, "DSA": 4.5, "BigData": 6.0}
    },
    {
        "id": "SV005",
        "name": "Hoàng Văn Em",
        "age": 22,
        "scores": {"SAD": 7.0, "DSA": 7.0, "BigData": 7.5}
    }
]


print(f"Số lượng sinh viên: {len(students)}")
````

````python
def calculate_average(scores):
    """
    Tính điểm trung bình từ dictionary điểm
    Args:
        scores (dict): Dictionary chứa điểm các môn
    Returns:
        float: Điểm trung bình
    """
    if not scores:
        return 0.0
    
    total_score = sum(scores.values())
    number_of_subjects = len(scores)
    
    return total_score / number_of_subjects

# 3. Hiển thị kết quả
print(f"Số lượng sinh viên: {len(students)}")
print("-" * 50)
print(f"{'ID':<7} | {'Họ và Tên':<15} | {'Điểm Trung Bình':<15}")
print("-" * 50)

for student in students:
    avg = calculate_average(student["scores"])
    # Hiển thị kết quả định dạng cột cho đẹp
    print(f"{student['id']:<7} | {student['name']:<15} | {avg:.2f}")

# 4. Kiểm tra thử với một bộ điểm riêng (test_scores)
test_scores = {"Math": 8.0, "English": 9.0}
print("-" * 50)
print(f"Điểm trung bình test_scores: {calculate_average(test_scores):.2f}")
````

## **IV. CÁC CHỨC NĂNG**

### 1: Hiển thị danh sách sinh viên

````python
def display_all_students():
    """
    Hiển thị danh sách tất cả sinh viên
    In ra: ID, Tên, Tuổi, Điểm trung bình
    """
    # code của bạn
    def display_all_students():
        print(f"\n{'ID':<7} | {'Họ và Tên':<20} | {'Tuổi':<5} | {'Điểm TB':<8}")
    print("-" * 50)
    for s in students:
        avg = calculate_average(s["scores"])
        print(f"{s['id']:<7} | {s['name']:<20} | {s['age']:<5} | {avg:.2f}")

# Test 
display_all_students()
````

### 2: Thêm sinh viên mới

````python
def add_new_student():
    """
    Thêm sinh viên mới
    - Tự động tạo ID theo format SV001, SV002...
    - Nhập: name, age
    - Nhập điểm 3 môn: SAD, DSA, BigData
    """
    print("\n--- THÊM SINH VIÊN MỚI ---")
    
    # code của bạn
    if not students:
        new_id = "SV001"
    else:
        last_id = students[-1]["id"]
        new_id_number = int(last_id[2:]) + 1
        new_id = f"SV{new_id_number:03d}"
    name = input("Nhập tên sinh viên: ")
    age = int(input("Nhập tuổi sinh viên: "))
    sad = float(input("Nhập điểm môn SAD: "))
    dsa = float(input("Nhập điểm môn DSA: "))
    bigdata = float(input("Nhập điểm môn BigData: "))

    # Thêm sinh viên mới vào danh sách
    students.append({
        "id": new_id,
        "name": name,
        "age": age,
        "scores": {
            "SAD": sad,
            "DSA": dsa,
            "BigData": bigdata
        }
    })

    print(f"Đã thêm sinh viên mới với ID: {new_id}")

add_new_student()
````

### 3: Cập nhật điểm

````python
def update_scores():
    """
    Cập nhật điểm sinh viên
    - Nhập ID sinh viên
    - Hiển thị điểm hiện tại
    - Cho phép cập nhật từng môn (Enter để bỏ qua)
    - Cập nhật dictionary scores
    """
    print("\n--- CẬP NHẬT ĐIỂM ---")
    
    # code của bạn
    student_id = input("Nhập ID sinh viên cần cập nhật điểm: ").strip().upper()
    for s in students:
        if s["id"] == student_id:
            print(f"Điểm hiện tại : {s['scores']}")
            for subject in s["scores"]:
                new_val = input(f"Nhập điểm mới cho {subject} (Enter để bỏ qua): ").strip()
                if new_val:
                    s["scores"][subject] = float(new_val)
            print(f"Đã cập nhật điểm cho sinh viên {s['name']}.")
            return
    print("Không tìm thấy sinh viên với ID đã nhập.")


update_scores()
````

### 4: Xóa sinh viên theo ID

````python
def delete_student(ID):
    """
    Xóa sinh viên theo ID 
    """
    print("\n--- XÓA SINH VIÊN ---")
    # code của bạn
    global students
    for s in students:
        if s["id"] == ID.strip().upper():
            students.remove(s)
            print(f"Đã xóa sinh viên với ID: {ID}")
            return
    print("Không tìm thấy sinh viên với ID cần xóa.")

    delete_student(ID)
````

### 5: Tìm sinh viên theo tên

````python
def search_by_name(name_search):
    """
    Tìm sinh viên theo tên
    - Không phân biệt hoa thường
    - Tìm kiếm gần đúng (chứa từ khóa)
    """
    print("\n--- TÌM SINH VIÊN THEO TÊN ---")
    # code của bạn
    results = [s for s in students if name_search.lower() in s["name"].lower()]
    if results:
        for s in results:
            print(f"ID: {s['id']} | Tên: {s['name']} | Điểm TB: {calculate_average(s['scores']):.2f}")
    else:
        print("Không tìm thấy kết quả phù hợp.")

    search_by_name(name_search)
````

### 6: Xếp hạng sinh viên

````python
def rank_students():
    """
    Xếp hạng sinh viên theo điểm trung bình
    - Tính điểm TB cho tất cả sinh viên
    - Sắp xếp theo điểm TB giảm dần
    - Sử dụng sorted() với key=lambda
    - Hiển thị bảng xếp hạng có số thứ tự
    """
    print("\n--- BẢNG XẾP HẠNG SINH VIÊN ---")
    
    # code của bạn
    sorted_list = sorted(students, key=lambda s: calculate_average(s["scores"]), reverse=True)
    print(f"{'STT':<4} | {'ID':<7} | {'Tên':<20} | {'Điểm TB':<8}")
    for i, s in enumerate(sorted_list, 1):
        avg = calculate_average(s["scores"])
        print(f"{i:<4} | {s['id']:<7} | {s['name']:<20} | {avg:.2f}")

rank_students()
````

### 7: Thống kê theo mức điểm

````python
def statistics_by_grade():
    """
    Thống kê sinh viên theo mức điểm
    - Giỏi: điểm TB >= 8.0
    - Khá: 6.5 <= điểm TB < 8.0
    - Trung bình: 5.0 <= điểm TB < 6.5
    - Yếu: điểm TB < 5.0
    - Đếm số sinh viên mỗi loại
    """
    print("\n--- THỐNG KÊ THEO MỨC ĐIỂM ---")
    
    # code của bạn
    stats = {"Giỏi": 0, "Khá": 0, "Trung bình": 0, "Yếu": 0}
    for s in students:
        avg = calculate_average(s["scores"])
        if avg >= 8.0: stats["Giỏi"] += 1
        elif avg >= 6.5: stats["Khá"] += 1
        elif avg >= 5.0: stats["Trung bình"] += 1
        else: stats["Yếu"] += 1
    
    for grade, count in stats.items():
        print(f"- {grade}: {count} sinh viên")


statistics_by_grade()
````

### 8: Tìm sinh viên điểm cao nhất/thấp nhất

````python
def find_top_bottom_students():
    """
    Tìm sinh viên có điểm cao nhất và thấp nhất
    - Sử dụng max()/min() với key
    - Hiển thị thông tin chi tiết
    """
    print("\n--- SINH VIÊN ĐIỂM CAO NHẤT VÀ THẤP NHẤT ---")

    # code của bạn
    if not students: return
    top = max(students, key=lambda s: calculate_average(s["scores"]))
    bot = min(students, key=lambda s: calculate_average(s["scores"]))
    
    print("\n--- SINH VIÊN ĐIỂM CAO NHẤT ---")
    print(f"ID: {top['id']} | Tên: {top['name']} | ĐTB: {calculate_average(top['scores']):.2f}")
    print("\n--- SINH VIÊN ĐIỂM THẤP NHẤT ---")
    print(f"ID: {bot['id']} | Tên: {bot['name']} | ĐTB: {calculate_average(bot['scores']):.2f}")
    

    find_top_bottom_students()
````

````python
def main_menu():
    while True:
        print("\n" + "="*50)
        print("     CHƯƠNG TRÌNH QUẢN LÝ SINH VIÊN VÀ ĐIỂM")
        print("="*50)
        print("1. Hiển thị danh sách sinh viên")
        print("2. Thêm sinh viên mới")
        print("3. Cập nhật điểm")
        print("4. Xóa sinh viên theo ID")
        print("5. Tìm sinh viên theo tên")
        print("6. Xếp hạng sinh viên")
        print("7. Thống kê theo mức điểm")
        print("8. Tìm sinh viên điểm cao nhất/thấp nhất")
        print("0. Thoát")
        print("="*50)
        
        choice = input("Chọn chức năng (0-8): ").strip()
        
        if choice == "1":
           # code của bạn 
           if choice == "1": display_all_students()
        elif choice == "2": add_new_student()
        elif choice == "3": update_scores()
        elif choice == "4": 
            id_del = input("Nhập ID cần xóa: ")
            delete_student(id_del)
        elif choice == "5":
            name = input("Nhập tên cần tìm: ")
            search_by_name(name)
        elif choice == "6": rank_students()
        elif choice == "7": statistics_by_grade()
        elif choice == "8": find_top_bottom_students()
        elif choice == "0":
            print("Cảm ơn bạn đã sử dụng chương trình!")
            break
        else:
            print("\n✗ Lựa chọn không hợp lệ! Vui lòng chọn lại.")

# Chạy chương trình
if __name__ == "__main__":
    main_menu()
````
